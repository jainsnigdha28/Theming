```python
from string import punctuation

punctuation = [p for p in punctuation]

review = df_new["Reviews1"] 
stop_words = stopwords.words('english') + stop_w
stop_words = [w for w in stop_words if w != '']
stop_words = set(stop_words)
  
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer('\w+')

from string import digits 
remove_digits = str.maketrans('', '', digits) 
# words = []
dic = {}
def process(r, key):
    print(r, key, )
    words = []
    tokenized = nltk.word_tokenize(str(r))
    try:
        for i in tokenized:
            i = i.translate(remove_digits)
            i = " ".join(tokenizer.tokenize(i))
            if i not in punctuation and i not in stop_words and ~i.isdigit() and i != '':
                words.append(i.lower().strip())

        if df_new['Theme1'][key] not in dic.keys():
            dic[df_new['Theme1'][key]] = []
        dic[df_new['Theme1'][key]] = dic[df_new['Theme1'][key]] + words
    except:
        print('Key Exception')
    return words


tokenized = [w for key, r in enumerate(review) for w in process(r, key)]
print(dic)
```

```python
# print(tokenized)
def word_count(words):
    counts = dict()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts

dictionary = word_count(tokenized)
print(dictionary)
```
