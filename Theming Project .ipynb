```python
df = pd.read_csv("SampleReviews.csv")
string = df.to_string()
```

```python
cleaningreviews = re.sub('[^a-zA-Z0-9\n]',' ',string)
```

```python
tokens = nltk.word_tokenize(cleaningreviews)
no_duplicates = list(set(tokens))
unique = [i.lower() for i in no_duplicates]

for i in unique: 
    if i == '!'or '#' or '%' or '&' or '.': 
        unique.remove(i) 


no_integers = [x for x in unique if not (x.isdigit() or x[0] == '-' and x[1:].isdigit())]   
type(no_integers)
```
