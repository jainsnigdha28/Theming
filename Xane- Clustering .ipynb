```python
df_new = df[pd.notnull(df['Reviews1'])]
#df_new

stop_words_file= open("stopwords.txt", "r")
stop_w = [line.strip() for line in open('stopwords.txt')]
```

```python
from string import punctuation

punctuation = [p for p in punctuation]

review = df["Reviews1"] #"Reviews2", "Reviews3"
stop_words = stopwords.words('english') + stop_w
stop_words = [w for w in stop_words if w != '']
stop_words = set(stop_words)
  
```

```python
def process(r):
    tokenized = nltk.word_tokenize(str(r))
    words = []
    for i in tokenized:
        if i not in punctuation and i not in stop_words and ~i.isdigit():
            words.append(i.lower().strip())
    return words


tokenized = [process(r) for r in review]
print(tokenized)
```

```python
words = []
for sublist in tokenized:
    for item in sublist:
        words.append(item)  
```

```python
def word_count(str):
    counts = dict()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts

dictionary = word_count(words)
```

```python
import operator
sorted_dictionary = dict(sorted(dictionary.items(), key=operator.itemgetter(1),reverse=True))
print(sorted_dictionary)
```
